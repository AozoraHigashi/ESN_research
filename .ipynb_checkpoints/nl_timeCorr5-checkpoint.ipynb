```python
""" Orstein-Uhlenbeck Process !!!!! """

import torch
import matplotlib.pyplot as plt
import ESN
import time
import numpy as np
import util
torch.set_default_device("cuda:0")
torch.set_default_dtype(torch.double)
# Parameters
Two,Ttrain = 2000,100000
N = 10
C = 0
rho = 0.6
sigma = 1
dim = 1
idWin = False
actf="tanh"
rseed = 0
N_d = int(N * dim)


esn = ESN.ESN_mult(N_d,sigma=sigma, uC=C,rho=rho, dim=dim,rseed=0,idWin=idWin)

fn = "nl_timeCorr_OU_v2"
## store setting
setting = {"input dim":dim,"Two":Two, "Ttrain":Ttrain,"sigma":sigma,"rseed" : rseed, 
            "Nodes":N_d,"uC":C,"rho":rho, "actf":actf,"identical Win":idWin,"input dist":"brownian motion"}
print("settings stored:",setting)
```

```python
fn_in = "100K_nl_normal"

## create target functions
if 1:
    x = torch.load('./experiments/inputs/'+fn_in+'_i.pt',weights_only=False)
    print("load complete")
    
t,y,dy = util.Ornstein_Uhlenbeck(x,3,0,3)

u = y.unsqueeze(0)
u_tld = x

```

```python
## run and washout
st = time.time()
Xwo = esn.run_washout(u, Two, actf=actf,bias=False)
print("runtime :",time.time()-st)
```

```python
if 0 :
    torch.save(u,f"./experiments/inputs/{fn}_i.pt")
    torch.save(Xwo,f"./experiments/datamatrices/{fn}_d.pt")
```

```python
#fig,ax = plt.subplots(figsize=(12,8))
T_len=10000
#T_len=-1
t=np.arange(Two+Ttrain)
#t=t.cpu().numpy()

for d in range(dim):
    plt.plot(t[-T_len:],u[d][-T_len:].cpu().numpy(),alpha=0.8)
plt.xlabel("t")
plt.ylabel(r'u(t)')
plt.title('input : Orstein-Uhlenbeck process')
plt.grid(True)
plt.show()


for d in range(dim):
    plt.plot(t[-T_len:],dy[-T_len:].cpu().numpy(),alpha=0.8)
plt.xlabel("t")
plt.ylabel(r'delta u(t)')
plt.title('delta u : Orstein-Uhlenbeck process')
plt.grid(True)
plt.show()

```
