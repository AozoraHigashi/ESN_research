```python
import torch
import matplotlib.pyplot as plt
import ESN
import time
import numpy as np
torch.set_default_device("cuda:0")
torch.set_default_dtype(torch.double)
# Parameters
Two,Ttrain = 2000,10000
N = 100
C = 0.9
dim = 2

N_d = int(N * dim)
maxtau = int(N_d * 2)


taus = np.arange(1,maxtau)

torch.manual_seed(0)


u1 = torch.rand(1, Two + Ttrain)*5
u2 = torch.rand(1, Two + Ttrain)
u = torch.vstack((u1[0],u1[0]+u2[0]))
u_tilde = torch.vstack((u1[0],u2[0]))

esn_idwin = ESN.ESN_mult(N_d, uC=C, dim=dim,rseed=0,idWin=True)
esn = ESN.ESN_mult(N_d, uC=C, dim=dim,rseed=0,idWin=False)
```

```python
""" NOT ID WIN  """
st = time.time()
Xwo = esn.run_washout(u, Two, actf="identity",bias=False)
print("runtime :",time.time()-st)

st = time.time()
#
raw,lin,rev,sur = ESN.MC_cSVD_asym(u, Xwo, maxtau,ret_all=True)
#
raw_tilde,lin_tilde,rev_tilde,sur_tilde = ESN.MC_cSVD_asym(u_tilde, Xwo, maxtau,ret_all=True)
mfs=rev
mfs_tilde = rev_tilde
print("calc time :",time.time()-st)


totMC = torch.sum(mfs).cpu().numpy()
totMC_tilde = torch.sum(mfs_tilde).cpu().numpy()

print(f"totMC:{totMC}\nsurrogate value: {sur.cpu().numpy()}")

print(f"totMC_tilde:{totMC_tilde}")
```

```python
""" ID WIN  """
st = time.time()
Xwo = esn_idwin.run_washout(u, Two, actf="identity",bias=False)
print("runtime :",time.time()-st)
st = time.time()
raw,lin,rev,sur = ESN.MC_cSVD_asym(u, Xwo, maxtau,ret_all=True)
raw_tilde,lin_tilde,rev_tilde,sur_tilde = ESN.MC_cSVD_asym(u_tilde, Xwo, maxtau,ret_all=True)
mfs_id=rev
mfs_tilde_id = rev_tilde
print("calc time :",time.time()-st)
totMC = torch.sum(mfs).cpu().numpy()
totMC_tilde = torch.sum(mfs_tilde).cpu().numpy()
print(f"totMC:{totMC}\nsurrogate value: {sur.cpu().numpy()}")
print(f"totMC_tilde:{totMC_tilde}")
```

```python
""" NOT ID WIN  """
fig,axes = plt.subplots(1,2,figsize=(16,6))
for d in range(dim):
    axes[0].plot(taus,mfs[d].cpu().numpy(),alpha=0.6,label="dim "+str(d))
    axes[1].plot(taus,mfs_tilde[d].cpu().numpy(),alpha=0.6,label="dim "+str(d)+" normalized")
    

axes[0].set_title(r'Ttr = %d, C=%.2f,$N=%d, MC=%.1f$'%(Ttrain,C,N_d,totMC))
axes[1].set_title(r'Ttr = %d, C=%.2f,$N=%d, \tilde{MC}=%.1f$'%(Ttrain,C,N_d,totMC_tilde))
axes[0].set_ylabel(r'Memory function $MC^{(k)}(\tau)$')
axes[1].set_ylabel(r'Memory function $\tilde{MC}^{(k)}(\tau)$')

for i in range(2):
    axes[i].set_ylim(-0.1,1.1)
    axes[i].set_xlabel("delay")
    axes[i].legend()
    axes[i].grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.95])
fig.suptitle("NOT identical Win", fontsize=16)
fig.show()


""" ID WIN  """
fig,axes = plt.subplots(1,2,figsize=(16,6))
for d in range(dim):
    axes[0].plot(taus,mfs_id[d].cpu().numpy(),alpha=0.6,label="dim "+str(d))
    axes[1].plot(taus,mfs_tilde_id[d].cpu().numpy(),alpha=0.6,label="dim "+str(d)+" normalized")
    

axes[0].set_title(r'Ttr = %d, C=%.2f,$N=%d, MC=%.1f$'%(Ttrain,C,N_d,totMC))
axes[1].set_title(r'Ttr = %d, C=%.2f,$N=%d, \tilde{MC}=%.1f$'%(Ttrain,C,N_d,totMC_tilde))
axes[0].set_ylabel(r'Memory function $MC^{(k)}(\tau)$')
axes[1].set_ylabel(r'Memory function $\tilde{MC}^{(k)}(\tau)$')

for i in range(2):
    axes[i].set_ylim(-0.1,1.1)
    axes[i].set_xlabel("delay")
    axes[i].legend()
    axes[i].grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.95])
fig.suptitle("identical Win", fontsize=16)
fig.show()
```
