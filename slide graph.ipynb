```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y=np.array([3,5,1,3,7,4,3,5])
num= len(y)
x=np.arange(1,len(y)+1)
delays=np.flip(x)
```

```python
plt.figure(figsize=(5,3))
err = np.random.normal(0,0.5, num)
ind = np.flip(np.argsort(np.abs(err)))
err = err[ind]
y_err=y+err
plt.plot(x,y,"bo-",label="input history")
plt.plot(x,y_err,"ro--",label="delay memory")

plt.xlabel("timestep")
plt.ylabel("pitch")
#plt.title("delay memory")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,y_err,"ro-",label="memory output")
plt.plot(9,1,"bo",label="next input")
plt.hlines(1,1,9,"b","--",alpha=0.5)

plt.xticks(np.append(x,9), np.append(delays, 0))
plt.xlabel("delay")
#plt.xlabel("timestep")

plt.ylabel("pitch")
#plt.title("delay memory")
plt.grid(True)
plt.legend()
plt.show()
```
