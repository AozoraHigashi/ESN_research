```python
import torch
import matplotlib.pyplot as plt
import ESN
import time
import numpy as np
torch.set_default_device("cuda:0")
torch.set_default_dtype(torch.double)

# Parameters
Two,Ttrain = 2000,100000
#Two,Ttrain = 2000,1000

N = 100
C = 0.95
maxdim = 50

trials = 100

dims = torch.arange(1, maxdim + 1)


#torch.manual_seed(0)

MC_N_trials = []
for i in np.arange(25,trials+1):
    #fn = f"MCN_tst_{i}.txt"
    fn = f"MCN_{i}.txt"
    totMCs = []
    totMC_N =[]
    for dim in dims:
        print( f"\r trial number {i}/{trials} : calculating {int(dim)}/{maxdim} dimensions...",end = "")
        seed = i*50+dim
        torch.manual_seed(seed)
        N_d = int(N * dim)
        #N_d = N
        maxtau = int(N_d / dim * 1.7)
        u_asym = torch.rand(dim, Two + Ttrain)
        u_sym = 2 * u_asym - 1

        esn = ESN.ESN_mult(N_d, uC=C, dim=dim)
        Xwo = esn.run_washout(u_sym, Two, actf="identity")
        mfs = ESN.MC_cSVD_old(u_sym, Xwo, maxtau)

        mc = torch.sum(mfs).cpu().numpy()
        totMCs.append(mc)
        totMC_N.append((mc/int(N_d)))
    # totMC_N : (maxdim,)
    print("\ntotMC_N:",totMC_N)
    np.savetxt(fn,totMC_N)
    MC_N_trials.append(totMC_N)
# MC_N_trials : (trials,maxdim)
np.savetxt("MCN_res.txt",MC_N_trials)
```
